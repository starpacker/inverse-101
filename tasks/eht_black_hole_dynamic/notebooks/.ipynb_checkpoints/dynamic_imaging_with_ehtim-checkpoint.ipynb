# Dynamic Imaging of EHT Black Hole with ehtim

This notebook demonstrates **dynamic (time-variable) imaging** of a synthetic black hole
using the `ehtim` library's `dynamical_imaging` module (Johnson et al. 2017).

We:
1. Generate a **ground-truth video** — a rotating crescent ring (M87*-like)
2. Simulate **noisy EHT observations** from the video using `Movie.observe()`
3. Reconstruct the video using three data-term strategies:

| Method | Data terms | Description |
|--------|-----------|-------------|
| **Visibility** | Complex visibilities | Standard but gain-sensitive |
| **Amp + Closure Phase** | Amplitudes + closure phases | Partially gain-robust |
| **Closure-only** | Closure phases + log closure amps | Fully gain-robust |

4. Compare reconstructions with ground truth (metrics + visualization)

```python
import os, sys, warnings, io
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec

warnings.filterwarnings('ignore')

import ehtim
import ehtim.const_def as ehc
from ehtim.imaging.dynamical_imaging import dynamical_imaging, split_obs

%matplotlib inline
plt.rcParams.update({'font.size': 10, 'figure.dpi': 120})
print(f'ehtim version: {ehtim.__version__}')
```

## 1. Ground-Truth Video: Rotating Crescent Ring

We create a synthetic M87*-like crescent ring whose bright spot **rotates** over time,
simulating the orbital motion of hot plasma around the black hole shadow.
The video spans 6 hours with one frame every 30 minutes (12 frames total).

```python
def make_ring_frame(N=64, ring_radius_frac=0.22, ring_width_frac=0.055,
                    asymmetry=0.5, asymmetry_angle_deg=220.0):
    """Create a single ring frame (normalized to sum=1)."""
    coords = np.linspace(-1.0, 1.0, N)
    xx, yy = np.meshgrid(coords, coords)
    r = np.sqrt(xx**2 + yy**2)
    theta = np.arctan2(yy, xx)
    phi0 = np.deg2rad(asymmetry_angle_deg)
    brightness = 1.0 + asymmetry * np.cos(theta - phi0)
    brightness = np.maximum(brightness, 0.0)
    ring = np.exp(-((r - ring_radius_frac)**2) / (2.0 * ring_width_frac**2)) * brightness
    ring = np.maximum(ring, 0.0)
    ring /= ring.sum()
    return ring


def make_rotating_ring_video(N=64, n_frames=12, rotation_deg=90.0, total_flux=0.6,
                              base_angle_deg=220.0):
    """Create a video of a rotating crescent ring.
    
    The bright spot rotates by `rotation_deg` over the full video duration.
    Returns frames in Jy/pixel and the angles used.
    """
    angles = np.linspace(base_angle_deg, base_angle_deg + rotation_deg, n_frames)
    frames = []
    for angle in angles:
        frame = make_ring_frame(N=N, asymmetry_angle_deg=angle)
        frames.append(frame * total_flux)  # Jy/pixel
    return frames, angles


# --- Parameters ---
N = 64
total_flux = 0.6       # Jy
pixel_size_uas = 2.0
psize = pixel_size_uas * ehc.RADPERUAS
n_frames = 12
rotation_deg = 90.0    # bright spot rotates 90 degrees over 6 hours

# Source coordinates (M87*)
ra_hours = 12.513728717168174
dec_deg  = 12.39112323919932

# Observation window
tstart = 0.0  # hours UTC
tstop  = 6.0

frames_jy, angles = make_rotating_ring_video(
    N=N, n_frames=n_frames, rotation_deg=rotation_deg, total_flux=total_flux
)
frame_times = np.linspace(tstart, tstop, n_frames)  # hours

print(f'Video: {n_frames} frames, {N}x{N} pixels')
print(f'Time span: {tstart:.1f} - {tstop:.1f} hours')
print(f'Bright spot rotation: {rotation_deg}° over {tstop - tstart:.1f} hours')
```

```python
# Display ground-truth video frames
fov = N * pixel_size_uas
hw = fov / 2

n_show = min(6, n_frames)
show_idx = np.linspace(0, n_frames - 1, n_show, dtype=int)

fig, axes = plt.subplots(1, n_show, figsize=(3.5 * n_show, 3.5))
vmax = max(f.max() for f in frames_jy)

for i, idx in enumerate(show_idx):
    ax = axes[i]
    im = ax.imshow(frames_jy[idx], cmap='afmhot', origin='lower',
                   extent=[-hw, hw, -hw, hw], vmin=0, vmax=vmax)
    ax.set_title(f't = {frame_times[idx]:.1f} h\n(angle={angles[idx]:.0f}°)', fontsize=9)
    ax.set_xlabel('μas')
    if i == 0:
        ax.set_ylabel('μas')

fig.suptitle('Ground-Truth Video: Rotating Crescent Ring', fontsize=12, y=1.02)
plt.tight_layout()
plt.show()
```

## 2. Build ehtim Movie & EHT Array

We wrap the ground-truth frames into an `ehtim.movie.Movie` object and load the EHT 2017 array.

```python
import tempfile

# --- Build ehtim Movie from frames ---
gt_movie = ehtim.movie.Movie(
    frames_jy, frame_times, psize, ra_hours, dec_deg,
    rf=230e9, source='M87', mjd=57849, interp='linear'
)
print(f'Movie: {gt_movie.nframes} frames, {gt_movie.xdim}x{gt_movie.ydim} pixels')
print(f'Time range: {gt_movie.start_hr:.2f} - {gt_movie.stop_hr:.2f} hours')

# --- EHT 2017 array ---
arr_txt = """ALMA   2225061.164  -5440057.370  -2481681.150    90
APEX   2225039.530  -5441197.630  -2479303.360  3500
JCMT  -5464584.680  -2493001.170   2150653.980  6000
SMA   -5464523.400  -2493147.080   2150611.750  4900
SMT   -1828796.200  -5054406.800   3427865.200  5000
LMT    -768713.964  -5988541.798   2063275.947   600
PV     5088967.900   -301681.600   3825015.800  1400
SPT          0.010         0.010  -6359609.700  5000
"""
tmpf = tempfile.NamedTemporaryFile(mode='w', suffix='.txt', delete=False)
tmpf.write(arr_txt); tmpf.close()
eht = ehtim.array.load_txt(tmpf.name)
os.unlink(tmpf.name)

print(f'\nEHT 2017 Array: {len(eht.tarr)} stations')
```

## 3. Simulate Noisy EHT Observations

We observe the movie with the EHT array, adding thermal noise.
The observation is split into per-scan chunks for dynamic imaging.

```python
# Observation parameters
tint = 10      # integration time (seconds)
tadv = 600     # scan cadence (seconds) — one scan every 10 min
bw   = 2e9     # bandwidth (Hz)

# Observe the movie with thermal noise
obs = gt_movie.observe(eht, tint=tint, tadv=tadv, tstart=tstart, tstop=tstop,
                       bw=bw, add_th_noise=True,
                       ampcal=True, phasecal=True,
                       ttype='direct', seed=42, verbose=False)
obs = obs.add_fractional_noise(0.01)  # 1% systematic noise floor

print(f'Total visibilities: {len(obs.data)}')
print(f'Time range: {obs.data["time"].min():.2f} - {obs.data["time"].max():.2f} hours')
print(f'Resolution: {obs.res() / ehc.RADPERUAS:.1f} μas')

# Split into per-scan observations
obs_list = split_obs(obs)
n_scans = len(obs_list)
print(f'Number of scans: {n_scans}')
print(f'Visibilities per scan: {[len(o.data) for o in obs_list]}')
```

```python
# Plot UV coverage colored by time
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# UV coverage
ax = axes[0]
u_gl = obs.data['u'] / 1e9
v_gl = obs.data['v'] / 1e9
times = obs.data['time']
sc = ax.scatter(u_gl, v_gl, s=4, c=times, cmap='viridis', alpha=0.7)
ax.scatter(-u_gl, -v_gl, s=4, c=times, cmap='viridis', alpha=0.3)
ax.set_xlabel('u (Gλ)')
ax.set_ylabel('v (Gλ)')
ax.set_title('UV-Coverage (colored by time)')
ax.set_aspect('equal')
plt.colorbar(sc, ax=ax, label='Time (hr)')

# Visibility amplitudes vs time
ax = axes[1]
ax.scatter(times, np.abs(obs.data['vis']), s=4, alpha=0.5, c='steelblue')
ax.set_xlabel('Time (hr)')
ax.set_ylabel('|V| (Jy)')
ax.set_title('Visibility Amplitudes vs Time')

plt.tight_layout()
plt.show()
```

## 4. Prepare Initial & Prior Images

For dynamic imaging, we need:
- A **prior image** (Gaussian blob) — same as for static imaging
- A list of **initial images** — one per scan, all starting from the same Gaussian

```python
# Gaussian prior/initial image
fov_rad = N * psize
im_prior = ehtim.image.make_empty(N, fov_rad, ra_hours, dec_deg,
                                  rf=230e9, source='M87', mjd=57849)
im_prior = im_prior.add_gauss(total_flux,
                              (20 * ehc.RADPERUAS, 20 * ehc.RADPERUAS, 0, 0, 0))

# Initial image list: one per scan
init_ims = [im_prior.copy() for _ in range(n_scans)]

print(f'Prior image: {N}x{N}, total flux = {im_prior.total_flux():.3f} Jy')
print(f'Initial images: {len(init_ims)} (one per scan)')
```

## 5. Dynamic Reconstruction with Three Methods

We use `ehtim.imaging.dynamical_imaging.dynamical_imaging()` with:

- **Temporal regularization** (`R_dt`): penalizes rapid changes between consecutive frames
- **Spatial regularizers**: Maximum Entropy (`gs`) + Total Variation (`tv2`)
- Three different **data terms** to compare gain robustness:
  1. `vis` — Complex visibilities
  2. `amp` + `cphase` — Amplitudes + closure phases
  3. `cphase` + `logcamp` — Closure-only

```python
def run_dynamic_imaging(obs_list, init_ims, prior, d1, d2=False, d3=False,
                        alpha_d1=100, alpha_d2=50, alpha_d3=50,
                        R_dt_alpha=10.0, maxit=100, label=''):
    """Run ehtim dynamical_imaging and return a list of reconstructed Images."""
    old_stdout = sys.stdout
    sys.stdout = io.StringIO()
    try:
        result = dynamical_imaging(
            obs_list, init_ims, prior,
            d1=d1, d2=d2, d3=d3,
            alpha_d1=alpha_d1, alpha_d2=alpha_d2, alpha_d3=alpha_d3,
            systematic_noise1=0.02, systematic_noise2=0.02, systematic_noise3=0.02,
            entropy1='gs', entropy2='tv2',
            alpha_s1=1.0, alpha_s2=1.0,
            R_dt={'alpha': R_dt_alpha, 'metric': 'SymKL', 'sigma_dt': 0.0, 'p': 2.0},
            R_dI={'alpha': 1.0, 'metric': 'SymKL', 'p': 2.0},
            alpha_flux=1.0,
            alpha_centroid=0.0,
            maxit=maxit,
            ttype='direct',
            processes=-1,
            clipfloor=0.0,
            Target_Dynamic_Range=10000.0,
        )
    finally:
        sys.stdout = old_stdout
    
    # Result can be a list of Images or a Movie
    if isinstance(result, list):
        return result
    else:
        return result.im_list()


# --- Method 1: Visibility ---
print('Reconstructing with Visibility data term...')
recon_vis = run_dynamic_imaging(
    obs_list, init_ims, im_prior,
    d1='vis', alpha_d1=100,
    R_dt_alpha=10.0, maxit=100, label='Vis'
)
print(f'  -> {len(recon_vis)} frames')

# --- Method 2: Amplitude + Closure Phase ---
print('Reconstructing with Amp + CPhase data terms...')
recon_amp_cp = run_dynamic_imaging(
    obs_list, init_ims, im_prior,
    d1='amp', d2='cphase', alpha_d1=100, alpha_d2=50,
    R_dt_alpha=10.0, maxit=100, label='Amp+CP'
)
print(f'  -> {len(recon_amp_cp)} frames')

# --- Method 3: Closure-only ---
print('Reconstructing with Closure-only data terms...')
recon_closure = run_dynamic_imaging(
    obs_list, init_ims, im_prior,
    d1='cphase', d2='logcamp', alpha_d1=50, alpha_d2=50,
    R_dt_alpha=10.0, maxit=100, label='Closure'
)
print(f'  -> {len(recon_closure)} frames')

print('\nAll reconstructions complete!')
```

## 6. Quantitative Comparison: Metrics

We compute per-frame **NRMSE** and **NCC** (normalized cross-correlation) between
each reconstruction and the ground-truth video.

```python
def compute_metrics(est, ref):
    """Compute NRMSE and NCC after flux-normalizing the estimate."""
    est_n = est * (ref.sum() / (est.sum() + 1e-30))
    nrmse = float(np.sqrt(np.mean((est_n - ref)**2))
                  / (np.sqrt(np.mean(ref**2)) + 1e-30))
    ncc = float(np.sum(est_n * ref)
               / (np.sqrt(np.sum(est_n**2)) * np.sqrt(np.sum(ref**2)) + 1e-30))
    return {'nrmse': nrmse, 'ncc': ncc}


def get_gt_frame_for_scan(scan_time, frame_times, frames_jy):
    """Get the closest ground-truth frame for a given scan time."""
    idx = np.argmin(np.abs(frame_times - scan_time))
    return frames_jy[idx], idx


# Compute scan times from observation list
scan_times = [np.mean(o.data['time']) for o in obs_list]

# Compute per-frame metrics for each method
methods = {
    'Visibility': recon_vis,
    'Amp + CPhase': recon_amp_cp,
    'Closure-only': recon_closure,
}

all_metrics = {}
for method_name, recon_list in methods.items():
    frame_metrics = []
    for i, recon_im in enumerate(recon_list):
        gt_frame, gt_idx = get_gt_frame_for_scan(scan_times[i], frame_times, frames_jy)
        recon_arr = recon_im.imarr(pol='I')
        m = compute_metrics(recon_arr, gt_frame)
        m['scan_time'] = scan_times[i]
        m['gt_idx'] = gt_idx
        frame_metrics.append(m)
    all_metrics[method_name] = frame_metrics

# Summary table: average metrics
print(f'{"Method":<20s} {"Avg NRMSE":>10s} {"Avg NCC":>10s}')
print('-' * 42)
for name, fmetrics in all_metrics.items():
    avg_nrmse = np.mean([m['nrmse'] for m in fmetrics])
    avg_ncc = np.mean([m['ncc'] for m in fmetrics])
    print(f'{name:<20s} {avg_nrmse:>10.4f} {avg_ncc:>10.4f}')
```

```python
# Plot per-frame metrics over time
fig, axes = plt.subplots(1, 2, figsize=(12, 4.5))

colors = {'Visibility': 'tomato', 'Amp + CPhase': 'steelblue', 'Closure-only': 'forestgreen'}

for name, fmetrics in all_metrics.items():
    ts = [m['scan_time'] for m in fmetrics]
    nrmses = [m['nrmse'] for m in fmetrics]
    nccs = [m['ncc'] for m in fmetrics]
    axes[0].plot(ts, nrmses, 'o-', label=name, color=colors[name], markersize=4)
    axes[1].plot(ts, nccs, 'o-', label=name, color=colors[name], markersize=4)

axes[0].set_xlabel('Scan Time (hr)')
axes[0].set_ylabel('NRMSE')
axes[0].set_title('Per-Frame NRMSE (lower is better)')
axes[0].legend(fontsize=8)
axes[0].grid(alpha=0.3)

axes[1].set_xlabel('Scan Time (hr)')
axes[1].set_ylabel('NCC')
axes[1].set_title('Per-Frame NCC (higher is better)')
axes[1].legend(fontsize=8)
axes[1].grid(alpha=0.3)

fig.suptitle('Reconstruction Quality Over Time', fontsize=12, y=1.02)
plt.tight_layout()
plt.show()
```

## 7. Visual Comparison: Ground Truth vs Reconstructions

We show selected frames from each method side-by-side with the ground truth.

```python
# Select frames to display (evenly spaced)
n_display = min(6, n_scans)
display_idx = np.linspace(0, n_scans - 1, n_display, dtype=int)

method_names = ['Visibility', 'Amp + CPhase', 'Closure-only']
method_recons = [recon_vis, recon_amp_cp, recon_closure]

n_rows = 1 + len(method_names)  # GT + 3 methods
fig, axes = plt.subplots(n_rows, n_display, figsize=(3.2 * n_display, 3.2 * n_rows))

vmax_gt = max(frames_jy[i].max() for i in range(n_frames))

for j, scan_idx in enumerate(display_idx):
    gt_frame, gt_idx = get_gt_frame_for_scan(scan_times[scan_idx], frame_times, frames_jy)
    t_hr = scan_times[scan_idx]
    
    # Row 0: Ground truth
    ax = axes[0, j]
    ax.imshow(gt_frame, cmap='afmhot', origin='lower',
              extent=[-hw, hw, -hw, hw], vmin=0, vmax=vmax_gt)
    ax.set_title(f't={t_hr:.1f}h', fontsize=8)
    if j == 0:
        ax.set_ylabel('Ground Truth', fontsize=9)
    ax.tick_params(labelsize=6)
    
    # Rows 1-3: Reconstructions
    for row, (mname, recon_list) in enumerate(zip(method_names, method_recons)):
        ax = axes[row + 1, j]
        recon_arr = recon_list[scan_idx].imarr(pol='I')
        ax.imshow(recon_arr, cmap='afmhot', origin='lower',
                  extent=[-hw, hw, -hw, hw], vmin=0, vmax=recon_arr.max())
        m = all_metrics[mname][scan_idx]
        ax.set_title(f'NCC={m["ncc"]:.2f}', fontsize=7)
        if j == 0:
            ax.set_ylabel(mname, fontsize=9)
        ax.tick_params(labelsize=6)

fig.suptitle('Dynamic Imaging: Ground Truth vs Reconstructions', fontsize=13, y=1.01)
plt.tight_layout()
plt.show()
```

## 8. Temporal Evolution Analysis

We visualize how well each method captures the **rotation** of the bright spot
by tracking the angle of the brightness peak over time.

```python
def measure_peak_angle(image_2d):
    """Measure the angle of the brightness peak relative to center."""
    N = image_2d.shape[0]
    coords = np.linspace(-1.0, 1.0, N)
    xx, yy = np.meshgrid(coords, coords)
    
    # Weighted centroid angle (using top 20% pixels)
    threshold = image_2d.max() * 0.5
    mask = image_2d > threshold
    if mask.sum() == 0:
        return np.nan
    
    weights = image_2d[mask]
    x_weighted = np.average(xx[mask], weights=weights)
    y_weighted = np.average(yy[mask], weights=weights)
    return np.rad2deg(np.arctan2(y_weighted, x_weighted))


# Ground truth angles
gt_peak_angles = []
for scan_idx in range(n_scans):
    gt_frame, _ = get_gt_frame_for_scan(scan_times[scan_idx], frame_times, frames_jy)
    gt_peak_angles.append(measure_peak_angle(gt_frame))

# Reconstructed angles
fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(scan_times, gt_peak_angles, 'k-o', label='Ground Truth', markersize=6, lw=2)

for mname, recon_list in zip(method_names, method_recons):
    recon_angles = []
    for scan_idx in range(n_scans):
        recon_arr = recon_list[scan_idx].imarr(pol='I')
        recon_angles.append(measure_peak_angle(recon_arr))
    ax.plot(scan_times, recon_angles, 'o--', label=mname, color=colors[mname],
            markersize=5, alpha=0.8)

ax.set_xlabel('Time (hr)')
ax.set_ylabel('Peak Brightness Angle (deg)')
ax.set_title('Temporal Evolution: Bright Spot Rotation Tracking')
ax.legend(fontsize=9)
ax.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

## 9. Effect of Temporal Regularization Strength

We compare the visibility-based reconstruction with different `R_dt` alpha values
to understand the trade-off between temporal smoothness and fidelity.

```python
# Test different temporal regularization strengths
R_dt_alphas = [0.1, 1.0, 10.0, 100.0]
recon_by_alpha = {}

for alpha in R_dt_alphas:
    print(f'  R_dt alpha = {alpha} ...')
    recon = run_dynamic_imaging(
        obs_list, init_ims, im_prior,
        d1='vis', alpha_d1=100,
        R_dt_alpha=alpha, maxit=100, label=f'alpha={alpha}'
    )
    recon_by_alpha[alpha] = recon

print('Done!')
```

```python
# Plot average metrics vs R_dt alpha
avg_nrmse_by_alpha = []
avg_ncc_by_alpha = []

for alpha in R_dt_alphas:
    recon_list = recon_by_alpha[alpha]
    nrmses, nccs = [], []
    for i, recon_im in enumerate(recon_list):
        gt_frame, _ = get_gt_frame_for_scan(scan_times[i], frame_times, frames_jy)
        m = compute_metrics(recon_im.imarr(pol='I'), gt_frame)
        nrmses.append(m['nrmse'])
        nccs.append(m['ncc'])
    avg_nrmse_by_alpha.append(np.mean(nrmses))
    avg_ncc_by_alpha.append(np.mean(nccs))

fig, axes = plt.subplots(1, 2, figsize=(11, 4.5))

axes[0].semilogx(R_dt_alphas, avg_nrmse_by_alpha, 'o-', color='tomato', lw=2)
axes[0].set_xlabel('R_dt alpha (temporal regularization)')
axes[0].set_ylabel('Average NRMSE')
axes[0].set_title('NRMSE vs Temporal Regularization')
axes[0].grid(alpha=0.3)

axes[1].semilogx(R_dt_alphas, avg_ncc_by_alpha, 'o-', color='steelblue', lw=2)
axes[1].set_xlabel('R_dt alpha (temporal regularization)')
axes[1].set_ylabel('Average NCC')
axes[1].set_title('NCC vs Temporal Regularization')
axes[1].grid(alpha=0.3)

plt.tight_layout()
plt.show()

# Print table
print(f'{"R_dt alpha":<12s} {"Avg NRMSE":>10s} {"Avg NCC":>10s}')
print('-' * 34)
for alpha, nrmse, ncc in zip(R_dt_alphas, avg_nrmse_by_alpha, avg_ncc_by_alpha):
    print(f'{alpha:<12.1f} {nrmse:>10.4f} {ncc:>10.4f}')
```

```python
# Visual comparison of temporal regularization effect (middle frame)
mid_idx = n_scans // 2
gt_mid, _ = get_gt_frame_for_scan(scan_times[mid_idx], frame_times, frames_jy)

fig, axes = plt.subplots(1, len(R_dt_alphas) + 1, figsize=(3.5 * (len(R_dt_alphas) + 1), 3.5))

# Ground truth
axes[0].imshow(gt_mid, cmap='afmhot', origin='lower', extent=[-hw, hw, -hw, hw])
axes[0].set_title(f'Ground Truth\nt={scan_times[mid_idx]:.1f}h', fontsize=9)
axes[0].set_ylabel('μas')

for i, alpha in enumerate(R_dt_alphas):
    recon_arr = recon_by_alpha[alpha][mid_idx].imarr(pol='I')
    m = compute_metrics(recon_arr, gt_mid)
    axes[i + 1].imshow(recon_arr, cmap='afmhot', origin='lower',
                       extent=[-hw, hw, -hw, hw], vmin=0, vmax=recon_arr.max())
    axes[i + 1].set_title(f'α_dt={alpha}\nNCC={m["ncc"]:.3f}', fontsize=9)

for ax in axes:
    ax.set_xlabel('μas')
    ax.tick_params(labelsize=7)

fig.suptitle('Effect of Temporal Regularization (Visibility method, middle frame)', fontsize=11, y=1.02)
plt.tight_layout()
plt.show()
```

## 10. Summary

This notebook demonstrated:

1. **Ground-truth video generation**: A rotating crescent ring simulating orbital motion of hot plasma
2. **EHT observation simulation**: Realistic uv-coverage and thermal noise via `ehtim.movie.Movie.observe()`
3. **Dynamic reconstruction** with `ehtim.imaging.dynamical_imaging.dynamical_imaging()`:
   - Three data-term strategies (visibility, amp+closure phase, closure-only)
   - Temporal regularization (`R_dt`) to enforce frame-to-frame continuity
4. **Quantitative evaluation**: Per-frame NRMSE and NCC metrics
5. **Temporal fidelity**: Tracking the bright spot rotation angle over time
6. **Regularization trade-off**: Effect of `R_dt` strength on reconstruction quality

### Key findings:
- Dynamic imaging recovers time-variable structure from sparse EHT data
- Temporal regularization is essential — too weak leads to noisy frames, too strong leads to over-smoothing
- Visibility-based reconstruction gives best quality with calibrated data
- Closure-only reconstruction is gain-robust but lower resolution
