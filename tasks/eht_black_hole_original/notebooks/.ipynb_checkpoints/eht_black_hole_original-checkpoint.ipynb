## 1. Physics Background

### Very Long Baseline Interferometry (VLBI)

VLBI uses an array of radio telescopes spread across the Earth to achieve angular resolution equivalent to a telescope the size of the planet. Each pair of telescopes (a **baseline**) measures one Fourier component of the sky brightness distribution:

$$V(u,v) = \iint I(l,m)\, e^{-2\pi i(ul + vm)}\, dl\, dm$$

where $(u,v)$ are baseline coordinates in wavelengths and $(l,m)$ are sky coordinates.

### The Calibration Problem

In practice, each station $i$ introduces a complex gain $g_i$ (from atmosphere, electronics, etc.), so the **observed** visibility on baseline $i$-$j$ is:

$$V_{ij}^{\rm obs} = g_i \, g_j^* \, V_{ij}^{\rm true} + \varepsilon_{ij}$$

Traditional imaging assumes gains are calibrated out. When calibration fails, visibility-based imaging produces artifacts.

### Closure Quantities

**Closure phases** (on triangles of 3 stations) and **closure amplitudes** (on quadrangles of 4 stations) are combinations where station-based gains **cancel exactly**:

- **Closure phase**: $\phi_{ijk} = \arg(V_{ij} V_{jk} V_{ki})$ &mdash; gains cancel in the product
- **Closure amplitude**: $\text{CA}_{ijkl} = |V_{ij}||V_{kl}| / (|V_{ik}||V_{jl}|)$ &mdash; gain amplitudes cancel in the ratio

Chael et al. (2018) showed that imaging using **only** closure quantities produces correct results even when gains are unknown or corrupted.

## 2. EHT Array & Synthetic Data

We simulate an observation with the 9-station EHT 2017 array. Station positions use the WGS84 ellipsoid via `astropy`, and UV coordinates are computed from proper Earth-rotation synthesis (Greenwich Mean Sidereal Time).

```python
# --- EHT station properties ---
print(f"{'Station':>6s}  {'Lat':>8s}  {'Lon':>9s}  {'Height':>7s}  {'SEFD (Jy)':>10s}")
print("-" * 50)
for name, props in EHT_TELESCOPES.items():
    print(f"{name:>6s}  {props['lat_deg']:>8.3f}  {props['lon_deg']:>9.3f}  {props['height_m']:>7.0f}  {props['sefd_jy']:>10.0f}")
```

```python
# --- Ground truth: M87*-like ring ---
N = 64
pixel_size_uas = 2.0
uas_to_rad = np.pi / (180 * 3600 * 1e6)
pixel_size_rad = pixel_size_uas * uas_to_rad

image_gt = make_ring_image(N=N)

fig, axes = plt.subplots(1, 2, figsize=(11, 4.5))
plot_image(image_gt, title="Ground Truth: M87*-like Ring", ax=axes[0],
           pixel_size_uas=pixel_size_uas, cmap="afmhot")

# Radial profile
coords = np.linspace(-1, 1, N)
xx, yy = np.meshgrid(coords, coords)
r = np.sqrt(xx**2 + yy**2)
r_bins = np.linspace(0, 0.6, 30)
profile = [image_gt[(r >= r_bins[i]) & (r < r_bins[i+1])].mean()
           for i in range(len(r_bins)-1)]
r_mid = 0.5 * (r_bins[:-1] + r_bins[1:])
axes[1].plot(r_mid * N * pixel_size_uas, profile, 'k-', lw=2)
axes[1].set_xlabel("Radius (\u03bcas)")
axes[1].set_ylabel("Mean brightness")
axes[1].set_title("Radial Brightness Profile")
axes[1].axvline(0.22 * N * pixel_size_uas, color='r', ls='--', alpha=0.5, label='ring radius')
axes[1].legend(fontsize=9)
plt.tight_layout()
```

```python
# --- UV coverage (astropy Earth-rotation synthesis) ---
uv_coords, station_ids, n_stations, timestamps, sefds, station_names = \
    simulate_eht_uv_coverage(
        source_ra_deg=187.7059, source_dec_deg=12.3911,
        obs_start_utc="2017-04-06T00:00:00",
        obs_duration_hours=6.0, n_time_steps=15, freq_ghz=230.0,
    )
M = len(uv_coords)
print(f"Stations: {n_stations}, Visibility samples: {M}")
print(f"Observation span: MJD {timestamps.min():.4f} \u2013 {timestamps.max():.4f}")

fig, axes = plt.subplots(1, 2, figsize=(13, 5))
plot_uv_coverage(uv_coords, ax=axes[0])

# Baseline length histogram
bl_len = np.sqrt(uv_coords[:, 0]**2 + uv_coords[:, 1]**2) / 1e9
axes[1].hist(bl_len, bins=25, color="steelblue", alpha=0.7, edgecolor='k')
axes[1].set_xlabel("Baseline length (G\u03bb)")
axes[1].set_ylabel("Count")
axes[1].set_title("Baseline Length Distribution")
plt.tight_layout()
```

### SEFD-Based Noise Model

The thermal noise on each baseline depends on the sensitivity of the two stations. The per-baseline noise standard deviation is:

$$\sigma_{ij} = \frac{1}{\eta} \frac{\sqrt{\text{SEFD}_i \cdot \text{SEFD}_j}}{\sqrt{2 \Delta\nu \, \tau_{\rm int}}}$$

where $\eta = 0.88$ (quantization efficiency), $\Delta\nu = 2$ GHz (bandwidth), and $\tau_{\rm int} = 10$ s (integration time).

ALMA (SEFD = 100 Jy) is ~50x more sensitive than other stations, so ALMA baselines have much lower noise.

```python
# --- Compute visibilities, add noise, corrupt with gains ---
rng = np.random.default_rng(SEED)

A = _build_measurement_matrix(uv_coords, N, pixel_size_rad)
vis_true = A @ image_gt.ravel()

# Per-baseline SEFD noise
noise_std_per_vis = compute_sefd_noise(station_ids, sefds)
noise = noise_std_per_vis * (
    rng.standard_normal(M) + 1j * rng.standard_normal(M)
) / np.sqrt(2)
vis_noisy = vis_true + noise

# Station gain corruption
gain_amp_error = 0.2
gain_phase_error_deg = 30.0
vis_corrupted, gains = apply_station_gains(
    vis_noisy, station_ids, n_stations,
    amp_error=gain_amp_error,
    phase_error_deg=gain_phase_error_deg,
    rng=rng,
)

# Show noise distribution
fig, axes = plt.subplots(1, 2, figsize=(13, 4.5))

axes[0].hist(noise_std_per_vis, bins=30, color='steelblue', alpha=0.7, edgecolor='k')
axes[0].set_xlabel("Thermal noise \u03c3 (Jy)")
axes[0].set_ylabel("Count")
axes[0].set_title("Per-Baseline Noise Distribution")
axes[0].axvline(np.median(noise_std_per_vis), color='r', ls='--',
                label=f"median = {np.median(noise_std_per_vis):.4f}")
axes[0].legend(fontsize=9)

# Gain amplitudes
axes[1].bar(range(n_stations), np.abs(gains), color='salmon', edgecolor='k')
axes[1].set_xticks(range(n_stations))
axes[1].set_xticklabels(station_names, rotation=45, ha='right')
axes[1].set_ylabel("|g|")
axes[1].set_title(f"Station Gain Amplitudes (\u00b1{gain_amp_error:.0%} error)")
axes[1].axhline(1.0, color='k', ls='--', lw=0.8)
plt.tight_layout()
```

## 3. Closure Quantities & Gain Invariance

The key property: closure phases and closure amplitudes computed from **corrupted** visibilities are identical to those from **true** visibilities, because station gains cancel.

```python
# --- Compute closure quantities ---
tri, tri_st = find_triangles(station_ids, n_stations)
quad, quad_st = find_quadrangles(station_ids, n_stations)
print(f"Triangles: {len(tri)}, Quadrangles: {len(quad)}")

# Closure phases: true vs corrupted
cp_true = compute_closure_phases(vis_noisy, tri, station_ids)
cp_corrupted = compute_closure_phases(vis_corrupted, tri, station_ids)

# Log closure amplitudes: true vs corrupted
lca_true = compute_log_closure_amplitudes(vis_noisy, quad)
lca_corrupted = compute_log_closure_amplitudes(vis_corrupted, quad)

fig, axes = plt.subplots(1, 2, figsize=(13, 5))

axes[0].scatter(np.degrees(cp_true), np.degrees(cp_corrupted), s=12, alpha=0.6, c='steelblue')
axes[0].plot([-180, 180], [-180, 180], 'r--', lw=1.5, label='y = x')
axes[0].set_xlabel("True Closure Phase (\u00b0)")
axes[0].set_ylabel("Corrupted Closure Phase (\u00b0)")
axes[0].set_title("Closure Phases: Gain Invariance")
axes[0].legend(fontsize=9)
axes[0].set_aspect('equal')

axes[1].scatter(lca_true, lca_corrupted, s=12, alpha=0.6, c='salmon')
lim = [min(lca_true.min(), lca_corrupted.min()), max(lca_true.max(), lca_corrupted.max())]
axes[1].plot(lim, lim, 'r--', lw=1.5, label='y = x')
axes[1].set_xlabel("True log CA")
axes[1].set_ylabel("Corrupted log CA")
axes[1].set_title("Log Closure Amplitudes: Gain Invariance")
axes[1].legend(fontsize=9)
axes[1].set_aspect('equal')

plt.tight_layout()
print(f"Max |CP difference|: {np.max(np.abs(cp_true - cp_corrupted)):.2e} rad")
print(f"Max |logCA difference|: {np.max(np.abs(lca_true - lca_corrupted)):.2e}")
```

The scatter plots confirm that closure quantities from corrupted visibilities lie exactly on the $y = x$ line, demonstrating perfect gain invariance.

## 4. Forward Model

The `ClosureForwardModel` wraps the DFT matrix $\mathbf{A}$ and provides:
- Standard forward/adjoint operators
- Closure phase and amplitude computation
- Analytic gradients for optimization (Eqs. 11-12 of Chael 2018)

```python
model = ClosureForwardModel(
    uv_coords=uv_coords, image_size=N, pixel_size_rad=pixel_size_rad,
    station_ids=station_ids, triangles=tri, quadrangles=quad,
)
print(model)

psf = model.psf()
dirty_true = model.dirty_image(vis_noisy)
dirty_corr = model.dirty_image(vis_corrupted)

fig, axes = plt.subplots(1, 4, figsize=(18, 4))
plot_image(image_gt, "Ground Truth", ax=axes[0], pixel_size_uas=pixel_size_uas)
plot_image(psf, "PSF (Dirty Beam)", ax=axes[1], pixel_size_uas=pixel_size_uas, cmap='RdBu_r')
plot_image(dirty_true, "Dirty Image (true vis)", ax=axes[2], pixel_size_uas=pixel_size_uas)
plot_image(dirty_corr, "Dirty Image (corrupted vis)", ax=axes[3], pixel_size_uas=pixel_size_uas)
plt.tight_layout()
```

## 5. Reconstruction: Closure vs Visibility

### Figure 4 equivalent

We compare five reconstruction methods:
1. **Vis (true)**: Visibility RML-TV with true (uncorrupted) visibilities &mdash; reference
2. **Vis (corrupted)**: Visibility RML-TV with gain-corrupted visibilities &mdash; **fails**
3. **CP-only (TV)**: Closure phase only + TV regularization
4. **CP+CA (TV)**: Closure phase + closure amplitude + TV
5. **CP+CA (MEM)**: Closure phase + closure amplitude + Maximum Entropy

```python
noise_std = np.median(noise_std_per_vis)

print("[1/5] Visibility RML-TV (true visibilities)...")
x_vis_true = VisibilityRMLSolver(
    regularizers=[(5e3, TVRegularizer())], n_iter=500,
).reconstruct(model, vis_noisy, noise_std)
print("done")
```

```python
print("[2/5] Visibility RML-TV (corrupted visibilities)...")
x_vis_corrupted = VisibilityRMLSolver(
    regularizers=[(5e3, TVRegularizer())], n_iter=500,
).reconstruct(model, vis_corrupted, noise_std)
print("done")
```

```python
# Closure data dict
sigma_cp = closure_phase_sigma(vis_corrupted, noise_std_per_vis, tri)
sigma_logca = closure_amplitude_sigma(vis_corrupted, noise_std_per_vis, quad)
closure_data = dict(
    cphases=cp_corrupted, log_camps=lca_corrupted,
    sigma_cp=sigma_cp, sigma_logca=sigma_logca,
    triangles=tri, quadrangles=quad,
)

print("[3/5] CP-only (TV)...")
x_cp_tv = ClosurePhaseOnlySolver(
    regularizers=[(1e2, TVRegularizer())],
    alpha_cp=100.0, n_iter=2000,
).reconstruct(model, closure_data, x0=x_vis_corrupted)
print("done")
```

```python
print("[4/5] CP+CA (TV)...")
x_cpca_tv = ClosurePhasePlusAmpSolver(
    regularizers=[(1e2, TVRegularizer())],
    alpha_cp=100.0, alpha_ca=100.0, n_iter=2000,
).reconstruct(model, closure_data, x0=x_vis_corrupted)

print("[5/5] CP+CA (MEM)...")
x_cpca_mem = ClosurePhasePlusAmpSolver(
    regularizers=[(5e3, MaxEntropyRegularizer())],
    alpha_cp=100.0, alpha_ca=100.0, n_iter=2000,
).reconstruct(model, closure_data, x0=x_vis_corrupted)
print("done")
```

```python
# --- Comparison panel (Figure 4 equivalent) ---
reconstructions = {
    "Vis (true)": x_vis_true,
    "Vis (corrupted)": x_vis_corrupted,
    "CP-only (TV)": x_cp_tv,
    "CP+CA (TV)": x_cpca_tv,
    "CP+CA (MEM)": x_cpca_mem,
}
metrics = {name: compute_metrics(img, image_gt) for name, img in reconstructions.items()}

fig = plot_comparison(reconstructions, ground_truth=image_gt,
                      pixel_size_uas=pixel_size_uas, metrics=metrics)
fig.suptitle("Figure 4 equivalent: Closure-Only vs Visibility Imaging",
             y=1.02, fontsize=13)
plt.tight_layout()
```

```python
# --- Metrics table ---
print_metrics_table(metrics)
```

## 6. Gain Robustness Sweep

### Figure 5 equivalent

We sweep the gain error level from 0% to 100% and compare visibility-based vs closure-based reconstruction quality. The key result: **closure methods remain stable** while visibility imaging degrades rapidly.

```python
gain_levels = [0.0, 0.05, 0.1, 0.2, 0.5, 1.0]
sweep_nrmse_vis = []
sweep_nrmse_clo = []
sweep_ncc_vis = []
sweep_ncc_clo = []

for amp_err in gain_levels:
    phase_err = amp_err * 150.0
    rng_sweep = np.random.default_rng(42)

    if amp_err > 0:
        vis_corr, _ = apply_station_gains(
            vis_noisy, station_ids, n_stations,
            amp_error=amp_err, phase_error_deg=phase_err, rng=rng_sweep,
        )
    else:
        vis_corr = vis_noisy.copy()

    # Visibility RML
    x_v = VisibilityRMLSolver(
        regularizers=[(5e3, TVRegularizer())], n_iter=500,
    ).reconstruct(model, vis_corr, noise_std)

    # Closure (CP+CA)
    cp_s = compute_closure_phases(vis_corr, tri, station_ids)
    lca_s = compute_log_closure_amplitudes(vis_corr, quad)
    sig_cp_s = closure_phase_sigma(vis_corr, noise_std_per_vis, tri)
    sig_lca_s = closure_amplitude_sigma(vis_corr, noise_std_per_vis, quad)
    cd_s = dict(cphases=cp_s, log_camps=lca_s, sigma_cp=sig_cp_s,
                sigma_logca=sig_lca_s, triangles=tri, quadrangles=quad)

    x_c = ClosurePhasePlusAmpSolver(
        regularizers=[(1e2, TVRegularizer())],
        alpha_cp=100.0, alpha_ca=100.0, n_iter=2000,
    ).reconstruct(model, cd_s, x0=x_v)

    m_v = compute_metrics(x_v, image_gt)
    m_c = compute_metrics(x_c, image_gt)
    sweep_nrmse_vis.append(m_v['nrmse'])
    sweep_nrmse_clo.append(m_c['nrmse'])
    sweep_ncc_vis.append(m_v['ncc'])
    sweep_ncc_clo.append(m_c['ncc'])

    print(f"  gain={amp_err:>5.0%}  Vis NRMSE={m_v['nrmse']:.4f}  Closure NRMSE={m_c['nrmse']:.4f}")
```

```python
# --- Gain robustness plot (Figure 5 equivalent) ---
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

pct = [g * 100 for g in gain_levels]

axes[0].plot(pct, sweep_nrmse_vis, 'o-', label="Visibility RML", c="tab:red", lw=2)
axes[0].plot(pct, sweep_nrmse_clo, 's-', label="Closure CP+CA", c="tab:blue", lw=2)
axes[0].set_xlabel("Gain Amplitude Error (%)")
axes[0].set_ylabel("NRMSE (lower is better)")
axes[0].set_title("Figure 5: Gain Robustness (NRMSE)")
axes[0].legend()
axes[0].grid(True, alpha=0.3)

axes[1].plot(pct, sweep_ncc_vis, 'o-', label="Visibility RML", c="tab:red", lw=2)
axes[1].plot(pct, sweep_ncc_clo, 's-', label="Closure CP+CA", c="tab:blue", lw=2)
axes[1].set_xlabel("Gain Amplitude Error (%)")
axes[1].set_ylabel("NCC (higher is better)")
axes[1].set_title("Figure 5: Gain Robustness (NCC)")
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
```

```python
# --- Closure quantity distributions ---
fig, axes = plt.subplots(1, 2, figsize=(13, 4.5))

axes[0].scatter(range(len(cp_corrupted)), np.degrees(cp_corrupted),
                s=6, alpha=0.6, c='steelblue')
axes[0].set_xlabel("Triangle index")
axes[0].set_ylabel("Closure Phase (\u00b0)")
axes[0].set_title("Closure Phases")
axes[0].axhline(0, color='k', ls='--', lw=0.5)
axes[0].set_ylim(-185, 185)

axes[1].scatter(range(len(lca_corrupted)), lca_corrupted,
                s=6, alpha=0.6, c='salmon')
axes[1].set_xlabel("Quadrangle index")
axes[1].set_ylabel("log CA")
axes[1].set_title("Log Closure Amplitudes")
axes[1].axhline(0, color='k', ls='--', lw=0.5)

plt.tight_layout()
```

## 7. Summary

| Aspect | Detail |
|--------|--------|
| **Forward model** | Linear DFT: $\mathbf{y} = \mathbf{A}\mathbf{x}$, extended to closure quantities |
| **Array** | 9-station EHT 2017 (astropy WGS84 positions, SEFD-based noise) |
| **Measurements** | 540 complex visibilities, corrupted by station gains |
| **Key insight** | Closure quantities cancel station gains; closure-only imaging is robust |
| **Methods** | Visibility RML, CP-only RML, CP+CA RML (TV and MEM regularizers) |
| **Result** | Closure methods produce correct reconstructions even with 20-50% gain errors |
