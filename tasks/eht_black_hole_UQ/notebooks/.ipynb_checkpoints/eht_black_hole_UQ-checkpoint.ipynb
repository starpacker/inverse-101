## 1. Background — Why Uncertainty Quantification?

### The Ill-Posed Inverse Problem

The EHT measures only ~1000 Fourier coefficients of a sky image with ~1000
pixels, but with huge gaps in the uv-plane. The inverse problem is severely
underdetermined: **many images are consistent with the data**.

Standard methods (CLEAN, RML) produce a single *point estimate* — they cannot
tell you which features are real and which are artefacts of regularization.

### Deep Probabilistic Imaging (DPI)

DPI (Sun & Bouman 2020) addresses this by learning the **full posterior
distribution** $p(\text{image} | \text{data})$ using a normalizing flow:

$$z \sim \mathcal{N}(0, I) \;\xrightarrow{G_\theta}\; x \sim q_\theta(x) \approx p(x | y)$$

The flow $G_\theta$ is trained to minimize the KL divergence from the posterior:

$$\theta^* = \arg\min_\theta\; \mathbb{E}_{z} \left[
\underbrace{\mathcal{L}_\text{data}(y, f(G_\theta(z)))}_\text{data fidelity}
+ \lambda \underbrace{\mathcal{R}(G_\theta(z))}_\text{image prior}
- \beta \underbrace{\log|\det \partial G_\theta / \partial z|}_\text{entropy}
\right]$$

The **entropy term** is critical: without it, the model collapses to a point
estimate (delta-function posterior).

### Closure Quantities

DPI uses **closure phases** and **log closure amplitudes** instead of raw
visibilities, because closures are immune to station-based gain calibration
errors:

- **Closure phase** (triangle): $\phi_{ijk} = \arg V_{ij} + \arg V_{jk} + \arg V_{ki}$
- **Log closure amp** (quadrangle): $\ln A_{ijkl} = \ln|V_{ij}| + \ln|V_{kl}| - \ln|V_{ik}| - \ln|V_{jl}|$

## 3. Forward Model — NUFFT & Closure Quantities

The forward model computes complex visibilities from an image using GPU-accelerated
Non-Uniform FFT (torchkbnufft), then derives closure phases and log closure
amplitudes via pre-computed index maps.

$$\text{image} \;\xrightarrow{\text{NUFFT}}\; V_{ij}
\;\xrightarrow{\text{index maps}}\; \phi_{ijk},\; \ln A_{ijkl}$$

Let's verify it works on the ground-truth image.

````python
# Build forward model
cphase_ind_torch = [torch.tensor(a, dtype=torch.long)
                    for a in closure_indices['cphase_ind_list']]
cphase_sign_torch = [torch.tensor(a, dtype=torch.float32)
                     for a in closure_indices['cphase_sign_list']]
camp_ind_torch = [torch.tensor(a, dtype=torch.long)
                  for a in closure_indices['camp_ind_list']]

forward_model = NUFFTForwardModel(
    npix, nufft_params['ktraj_vis'], nufft_params['pulsefac_vis'],
    cphase_ind_torch, cphase_sign_torch, camp_ind_torch,
    device=torch.device('cpu'),
)

# Forward pass on ground truth
gt_tensor = torch.tensor(ground_truth, dtype=torch.float32).unsqueeze(0)
vis, visamp, cphase, logcamp = forward_model(gt_tensor)

print(f'Visibilities shape  : {vis.shape}   (batch, [real;imag], M)')
print(f'Vis amplitudes      : {visamp.shape}')
print(f'Closure phases      : {cphase.shape}')
print(f'Log closure amps    : {logcamp.shape}')
````

````python
# Visualize forward model outputs
fig, axes = plt.subplots(1, 3, figsize=(14, 3.5))

axes[0].plot(visamp[0].detach().numpy(), '.', ms=2, alpha=0.5)
axes[0].set_xlabel('Baseline index')
axes[0].set_ylabel('Amplitude')
axes[0].set_title('Visibility Amplitudes')

axes[1].plot(cphase[0].detach().numpy(), '.', ms=2, alpha=0.5, color='coral')
axes[1].set_xlabel('Triangle index')
axes[1].set_ylabel('Degrees')
axes[1].set_title('Closure Phases')

axes[2].plot(logcamp[0].detach().numpy(), '.', ms=2, alpha=0.5, color='green')
axes[2].set_xlabel('Quadrangle index')
axes[2].set_ylabel('Log ratio')
axes[2].set_title('Log Closure Amplitudes')

for ax in axes:
    ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
````

## 4. Real-NVP Normalizing Flow Architecture

The generative model is a **Real-NVP** normalizing flow:

```
z ~ N(0, I)  →  [Flow_1] → perm → [Flow_2] → perm → ... → [Flow_K] → x_raw
                                                                         ↓
                                                          Softplus × scale → x ≥ 0
```

Each **Flow block** contains:
```
ActNorm → AffineCoupling → Reverse → ActNorm → AffineCoupling → Reverse
```

The **AffineCoupling** layer splits the input into two halves $(a, b)$:
- Apply a neural network to $a$: $(\log s, t) = \text{Net}(a)$
- Transform $b$: $b' = (b + t) \cdot \exp(\tanh(\log s))$
- The Jacobian is triangular → cheap log-det: $\sum \tanh(\log s)$

Let's verify the flow is invertible:

````python
# Small test: verify flow invertibility
ndim = npix * npix  # 1024
n_flow_test = 4

flow = RealNVP(ndim, n_flow=n_flow_test, affine=True, seqfrac=4)
flow.eval()

torch.manual_seed(123)
z_test = torch.randn(4, ndim)

# z → reverse → x
with torch.no_grad():
    x_gen, logdet_rev = flow.reverse(z_test)
    # x → forward → z'
    z_recon, logdet_fwd = flow.forward(x_gen)

error = torch.abs(z_recon - z_test).max().item()
print(f'Flow blocks        : {n_flow_test}')
print(f'Coupling layers    : {n_flow_test * 2} (2 per block)')
print(f'Dimension          : {ndim}')
print(f'Parameters         : {sum(p.numel() for p in flow.parameters()):,}')
print(f'Invertibility error: {error:.2e}  (should be < 1e-3)')
print(f'logdet check       : fwd + rev = {(logdet_fwd + logdet_rev).abs().max().item():.2e}')
````

````python
# Visualize what random flow outputs look like (before training)
with torch.no_grad():
    z_rand = torch.randn(8, ndim)
    x_rand, _ = flow.reverse(z_rand)
    # Apply softplus for positivity
    x_rand_img = torch.nn.Softplus()(x_rand.reshape(-1, npix, npix))

fig, axes = plt.subplots(2, 4, figsize=(14, 6))
for i, ax in enumerate(axes.flatten()):
    ax.imshow(x_rand_img[i].numpy(), origin='lower', cmap='afmhot')
    ax.set_title(f'Random sample {i+1}')
    ax.axis('off')
plt.suptitle('Untrained flow outputs (random noise)', fontsize=12)
plt.tight_layout()
plt.show()
````

## 5. DPI Training

Now we train the normalizing flow to approximate the posterior.

**Training loop** (per epoch):
1. Sample $z \sim \mathcal{N}(0, I)$ (batch of 32)
2. Generate images: $x = \text{Softplus}(G_\theta(z)) \times \text{scale}$
3. Forward model: compute closure phases and log closure amplitudes
4. Loss = data fidelity + image priors - entropy (log-det)
5. Adam optimizer + gradient clipping

**Note**: Full training takes ~30,000 epochs on GPU. For this tutorial we use
a reduced number of epochs. Set `N_EPOCH` higher for production results.

````python
# ── Training configuration ──────────────────────────────────────────────
# Reduce epochs for the tutorial (set to 30000 for full results)
N_EPOCH = 1000

solver = DPISolver(
    npix=npix,
    n_flow=metadata['n_flow'],        # 16 flow blocks
    seqfrac=metadata.get('seqfrac', 4),
    n_epoch=N_EPOCH,
    batch_size=metadata['batch_size'], # 32
    lr=metadata['lr'],                 # 1e-4
    logdet_weight=metadata['logdet_weight'],
    l1_weight=metadata['l1_weight'],
    tsv_weight=metadata['tsv_weight'],
    flux_weight=metadata['flux_weight'],
    center_weight=metadata['center_weight'],
    mem_weight=metadata['mem_weight'],
    grad_clip=metadata['grad_clip'],
    device=device,
)

print(f'Training DPI with {N_EPOCH} epochs on {device}...')
print(f'  n_flow={metadata["n_flow"]}, batch_size={metadata["batch_size"]}, lr={metadata["lr"]}')
````

````python
%%time
# Train the model
result = solver.reconstruct(
    obs_data, closure_indices, nufft_params, prior_image, flux_const
)
````

````python
# Plot training loss curves
plot_loss_curves(result['loss_history'])
plt.show()

losses = result['loss_history']['total']
print(f'Initial loss : {np.mean(losses[:10]):.2f}')
print(f'Final loss   : {np.mean(losses[-10:]):.2f}')
````

## 6. Posterior Sampling & Uncertainty Maps

After training, we can draw posterior samples **instantly** by pushing random
Gaussian vectors through the trained flow. No MCMC or iterative optimization
needed!

From the samples we compute:
- **Posterior mean**: best point estimate
- **Posterior std**: pixel-wise uncertainty map
- **Individual samples**: shows the range of plausible images

````python
# Draw posterior samples
N_SAMPLES = 500
print(f'Drawing {N_SAMPLES} posterior samples...')
posterior = solver.posterior_statistics(n_samples=N_SAMPLES)

print(f'Posterior mean peak  : {posterior["mean"].max():.6f}')
print(f'Posterior mean flux  : {posterior["mean"].sum():.4f}')
print(f'Mean uncertainty     : {posterior["std"].mean():.6f}')
````

````python
# Posterior summary: Ground Truth | Mean | Std | Samples
plot_posterior_summary(
    posterior['mean'], posterior['std'], posterior['samples'],
    ground_truth=ground_truth,
    pixel_size_uas=pixel_size_uas,
)
plt.suptitle('DPI Posterior Summary', y=1.02, fontsize=12, fontweight='bold')
plt.show()
````

````python
# More posterior samples — each is a plausible black hole image
plot_posterior_samples(posterior['samples'], n_show=8,
                       pixel_size_uas=pixel_size_uas)
plt.suptitle('Individual Posterior Samples', y=1.02, fontsize=12)
plt.show()
````

````python
# Pixel-wise uncertainty analysis
fig, axes = plt.subplots(1, 3, figsize=(14, 4))

# Posterior std map
im0 = axes[0].imshow(posterior['std'], origin='lower', cmap='viridis')
axes[0].set_title('Posterior Std (uncertainty)')
plt.colorbar(im0, ax=axes[0], fraction=0.046)

# Coefficient of variation (std / mean)
cv = posterior['std'] / (posterior['mean'] + 1e-10)
im1 = axes[1].imshow(cv, origin='lower', cmap='plasma', vmax=np.percentile(cv, 95))
axes[1].set_title('Coefficient of Variation (std/mean)')
plt.colorbar(im1, ax=axes[1], fraction=0.046)

# Histogram of pixel uncertainties
axes[2].hist(posterior['std'].flatten(), bins=50, color='steelblue',
             edgecolor='white', alpha=0.8)
axes[2].set_xlabel('Posterior Std')
axes[2].set_ylabel('Pixel count')
axes[2].set_title('Uncertainty Distribution')
axes[2].axvline(posterior['std'].mean(), color='red', ls='--',
                label=f'Mean = {posterior["std"].mean():.4f}')
axes[2].legend()

plt.tight_layout()
plt.show()
````

### 7.2 Radial Profile with Uncertainty Bands

````python
# Azimuthally-averaged radial profiles from posterior samples
center = npix // 2
y_grid, x_grid = np.mgrid[0:npix, 0:npix]
r_grid = np.sqrt((x_grid - center + 0.5)**2 + (y_grid - center + 0.5)**2)
r_max = npix // 2
r_bins = np.arange(0, r_max + 1)

def radial_profile(image, r_grid, r_bins):
    profile = np.zeros(len(r_bins) - 1)
    for i in range(len(r_bins) - 1):
        mask = (r_grid >= r_bins[i]) & (r_grid < r_bins[i+1])
        if mask.sum() > 0:
            profile[i] = image[mask].mean()
    return profile

# Compute profiles for all samples
n_prof = min(200, posterior['samples'].shape[0])
profiles = np.array([radial_profile(posterior['samples'][i], r_grid, r_bins)
                     for i in range(n_prof)])
r_uas = (r_bins[:-1] + 0.5) * pixel_size_uas

gt_profile = radial_profile(ground_truth, r_grid, r_bins)
mean_profile = np.mean(profiles, axis=0)
std_profile = np.std(profiles, axis=0)

fig, ax = plt.subplots(figsize=(8, 4.5))
ax.plot(r_uas, gt_profile, 'k-', lw=2.5, label='Ground Truth', zorder=5)
ax.plot(r_uas, mean_profile, 'r-', lw=2, label='Posterior Mean')
ax.fill_between(r_uas,
                mean_profile - 2*std_profile,
                mean_profile + 2*std_profile,
                alpha=0.2, color='red', label='±2σ band')
ax.fill_between(r_uas,
                mean_profile - std_profile,
                mean_profile + std_profile,
                alpha=0.3, color='red', label='±1σ band')
ax.set_xlabel('Radius (μas)')
ax.set_ylabel('Brightness')
ax.set_title('Azimuthally-Averaged Radial Profile with Uncertainty')
ax.legend()
ax.grid(True, alpha=0.3)
ax.set_xlim(0, r_uas[-1])
plt.tight_layout()
plt.show()
````

### 7.3 Pixel Correlation Analysis

````python
# Pixel-pixel correlation from posterior samples
# Pick a reference pixel on the bright ring
samples_flat = posterior['samples'].reshape(posterior['samples'].shape[0], -1)

# Find the brightest pixel in posterior mean
ref_idx = np.argmax(posterior['mean'])
ref_y, ref_x = np.unravel_index(ref_idx, (npix, npix))

# Compute correlation with the reference pixel
ref_samples = samples_flat[:, ref_idx]
corr_map = np.zeros(npix * npix)
for j in range(npix * npix):
    r = np.corrcoef(ref_samples, samples_flat[:, j])[0, 1]
    corr_map[j] = r
corr_map = corr_map.reshape(npix, npix)

fig, axes = plt.subplots(1, 2, figsize=(10, 4))
plot_image(posterior['mean'], ax=axes[0], title='Posterior Mean',
           pixel_size_uas=pixel_size_uas)
axes[0].plot(ref_x, ref_y, 'w+', ms=12, mew=2)
axes[0].annotate('ref pixel', (ref_x, ref_y), color='white',
                 fontsize=9, xytext=(5, 5), textcoords='offset points')

im = axes[1].imshow(corr_map, origin='lower', cmap='RdBu_r',
                     vmin=-1, vmax=1)
axes[1].set_title('Correlation with Reference Pixel')
axes[1].plot(ref_x, ref_y, 'k+', ms=12, mew=2)
plt.colorbar(im, ax=axes[1], fraction=0.046)

plt.tight_layout()
plt.show()
````

## Summary

| Aspect | Detail |
|--------|--------|
| **Method** | Deep Probabilistic Imaging (Real-NVP normalizing flow) |
| **Data products** | Closure phases + log closure amplitudes (gain-invariant) |
| **Architecture** | 16 flow blocks × 2 coupling layers = 32 affine couplings |
| **Training** | KL divergence minimization: data + prior - entropy |
| **Output** | Full posterior distribution → mean, std, samples |
| **Key advantage** | Instant posterior sampling (no MCMC), principled UQ |
| **Critical term** | Entropy (log-det) — prevents posterior collapse |

**Reference:** Sun, H. & Bouman, K.L. (2020). *Deep Probabilistic Imaging*. arXiv:2010.14462
